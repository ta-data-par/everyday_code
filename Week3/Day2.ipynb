```python
url='https://en.wikipedia.org/wiki/List_of_sovereign_states'
response=requests.get(url).content
soup=BeautifulSoup(response,'lxml')

table=soup.find_all('table',{'class':'wikitable'})
table[0]
```

```python
rows=table[0].find_all('tr')
rows=[row.text.strip().split('\n\n') for row in rows]

def restr(x):
    if len(x)>4:
        x[3]=='\n'.join(x[3:])
        del x[4:]
    elif len(x)==3:
        x.append("")
    return x
rows=[restr(x) for x in rows]
rows
```

```python
df=pd.DataFrame([len(x) for x in rows])
df[0].value_counts()
```

```python
colnames = rows[0]
data = rows[1:]

df = pd.DataFrame(data, columns=colnames)
df
```

```python
def output(url):
    r=requests.get(url)
    if r.status_code<300:
        print('alles gut')
    elif 500>r.status_code>=400:
        print('forbidden')
    else:
        print('server connection error')
    return r

r=output('http://youtube.com')

r.history
```

```python
import asyncio

urls = [
    'https://s3-eu-west-1.amazonaws.com/ih-materials/uploads/data-static/documents/breakfast.jpg',
    'https://s3-eu-west-1.amazonaws.com/ih-materials/uploads/data-static/documents/forbidden',
    'https://s3-eu-west-1.amazonaws.com/ih-materials/uploads/data-static/documents/the-html5-breakfast-site.html'
]

async def main():
    loop = asyncio.get_event_loop()
    futures = [loop.run_in_executor(None, requests.get, url) for url in urls]
    for response in await asyncio.gather(*futures):
        print(response.status_code)
```
