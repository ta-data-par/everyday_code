```python
import numpy as np
```

```python
def bayes_rule(priors,likelihood):
    marg=sum(np.multiply(priors,likelihood))
    post=np.divide(np.multiply(priors,likelihood),marg)
    return post
```

```python
probas=[1/3,1/3,1/3]
likelihood=[0.4,0.6,0.3]
bayes_rule(probas,likelihood)
```

```python
bayes_rule(probas,[])
```

# Online Store Example

```python
import pandas as pd
```

```python
n_draws=1000000000
prior=pd.Series(np.random.uniform(0,1,size=n_draws))
```

```python
prior.hist()
```

```python
def generative_model(param):
    result=np.random.binomial(50,param)
    return result
```

```python
observed=10
np.random.binomial(50,[0.1,0.4])
```

```python
sim_data=generative_model(prior)
```

```python
sim_data
```

```python
np.std([5.6,28.6,32.3,23.9,31.1,40.3,31.8,49.7,57.6,57.5,55.4,84.4,85,75,87,93.9,48.5,122.6,132.7,121.4])
```

```python
np.std(range(0,39,2))
```

```python
posterior=prior[list(map(lambda x: x==observed,sim_data))]
posterior
```

```python
posterior.hist()
```

```python
posterior.describe()
```

```python
print(posterior.quantile(.05),'|',posterior.quantile(.95))
```

## Maximum Likelihood Estimator

```python
rounded=posterior.round(2)
mode=rounded.mode()[0]
mode
```

```python
probability=list(rounded).count(mode)/len(rounded)
print('MLE = ',mode,"|",probability)
```
