```python
square=lambda x:x**2

square(5)
```

```python
def square_func(x):
    return x**2

square_func(5)
```

```python
summ=lambda a,b:a+b

summ('a','b')
```

```python
div= lambda x,y: print(x/y) if (y!=0) else print('hasdhas')

div(10,3)
div(10,0)

```

```python
def generate_range(lower):
    return lambda upper: range(lower,upper)

custom_range=generate_range(2)
print(custom_range(5))
```

```python
[x for x in custom_range(10)]
```

```python
[[x for x in custom_range(i)] for i in range(3,6)]
```

```python
[square(x) for x in range(6)]
```

```python
string=['Metallica','AC/DC','Queen','System-of-a-Down']
[(lambda x: x.replace('-',' '))(x) for x in string]
```
