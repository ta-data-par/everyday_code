# DataFrame Calculations

```python
import numpy as np
import pandas as pd

animals=pd.read_csv('https://s3-eu-west-1.amazonaws.com/ih-materials/uploads/data-static/data/module-2/animals.csv')

animals.head(3)
```

```python
animals.bodywt=animals.bodywt*1000
```

```python
animals.head(3)
```

```python
animals['brainratio']=animals.brainwt/animals.bodywt
animals.head()
```

```python
animals['reversecheck']=animals.bodywt/animals.brainwt[animals.brainwt>50]
animals
```

```python
animals['jazz']=np.where(animals.brainwt>50,animals.bodywt/animals.brainwt,0)
animals
```

```python
np.where(animals.brainwt>50,animals.bodywt/animals.brainwt,0)
```

```python
animals['average']=animals.mean(axis=1)
animals
```
